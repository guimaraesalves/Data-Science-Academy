# Stack Plots

```python
dias = [1, 2, 3, 4, 5]
```

```python
dormir = [7, 8, 6, 77, 7]
comer = [2, 3, 4, 5,3]
trabalhar = [7,8,7,2,2]
passear = [8,5,7,8,13]
```

# Pie Chart

```python
fatias = [7,2,2,13]
atividades = ['dormir', 'comer', 'trabelhar', 'pessear']
colunas = ['c', 'm', 'r', 'k']
```

```python
plt.pie(fatias, labels = atividades, colors = colunas, startangle = 90, shadow = True, explode = (0, 0.1,0,0))
plt.show()
```

# Pylab

```python
# Visualizando os gráficos dentro do Jupyter Notebook
# o Pylab combina funcionalidades do pyplot com funcionalidade do Numpy
from pylab import *
%matplotlib inline
```

```python
x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

# Definindo os eixos
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha');
```

```python
# Gráficos com 2 figuras
X = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # eixos da figura principal
axes2 = fig.add_axes([0.2, 0.5,0.4, 0.3]) # eixos da figura secundária


# Figura principal
axes1.plot(y, x, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes.set_title('Figura Principal')

# Figura secundária
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes.set_title('Figura Secundária');
```

```python
# Gráficos em Paralelo
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Titulo')
    
fig.tight_layout()
```

# Gráficos a partir do Numpy

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```python
# Plot e Scatter
fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(np.random.randn(50), color='red')

ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```python
# plots diversos
_, ax = plt.subplots(2,3)

ax[0,1].plot(np.random.randn(50), color = 'green', linestyle = '-')
ax[1,0].hist(np.random.randn(50))
ax[1,2].scatter(np.arange(50), np.random.randn(50), color ='red')
plt.show()
```

```python
# Controle dos eixos
fig, axes = plt.subplots(1,3, figsize = (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Eixos com range padrão")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Eixos menores")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Eixos customizados")
```

```python
# Escala
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Escala Padrão")

axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Escala Logaritmica (y)");
```

```python
# Grid
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# Grid padrão
axes[0].plot(x, x**2, x, x**3, lw = 2)
axes[0].grid(True)

#Grid customizado
axes[1].plot(x, x**2, x, x**3, lw = 2)
axes[1].grid(color = 'b', alpha = 0.5, linestyle = 'dashed', linewidth = 0.5)
```

```python
# Gráfico de Linhas Gêmeas
fig, ax1 = plt.subplots()

ax1.plot(x, x*82, lw=2, color='blue')
ax1.set_ylabel('Area', fontsize=18, color='blue')
for  label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color='red')
ax2.set_ylabel("Volume", fontsize=18, color='red')
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
# Diferentes estilos de Plots
xx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*randn(len(xx)))
axes[0].set_title('Scatter')

axes[1].step(n, n**2, lw=2)
axes[1].set_title('Step')

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title('Barras')

axes[3].fill_between(x, x**3, color="green", alpha=0.5);
axes[3].set_title('fill_between');
```
